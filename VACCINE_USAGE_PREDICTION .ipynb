# Model Building

```python
x=combined_data.iloc[:,0:-1]
y=combined_data.iloc[:,[-1]]
```

```python
# Lets build logistic regression model using statsmodel
num_features= x.copy()
num_features=sm.add_constant(num_features) # for beta not value
```

```python
# Data Partition
x_train,x_test,y_train,y_test=train_test_split(num_features,y,train_size=0.70,random_state=100)
print(x_train.shape)
print(y_train.shape)
```

```python
logit=sm.Logit(y_train,x_train)
lgf=logit.fit()
```

```python
print(lgf.summary())
```

```python
# Drop all the columns which has p- value > 0.05

cols=x_train.columns.tolist()
# setting an initial max P-value
max_p_value =1
while len(cols) > 0:
    # defining the train set
    x_train = x_train[cols]
    # fitting the model
    model = sm.Logit(y_train,x_train).fit()  
    # getting the p-values and the maximum p-value
    p_values = model.pvalues
    max_p_value = max(p_values)
    # name of the variable with maximum p-value
    feature_with_p_max= p_values.idxmax()
    if max_p_value > 0.05:
        cols.remove(feature_with_p_max)
        
    else:
        break
        
        
selected_features = cols
print(selected_features)
```

```python
x_train=x_train[selected_features]
x_test=x_test[selected_features]
```

```python
logit=sm.Logit(y_train,x_train)
lgf=logit.fit()
```

```python
print(lgf.summary())
```

###  Prediction On Train & Test Data

```python
# Prediction on train data
train=pd.concat([x_train,y_train],axis=1)
train['Probability']=lgf.predict(x_train)
train.head()
```

```python
train['Predicted']=np.where(train['Probability'] >= 0.7,1,0)  #thershold  >0.7
train.sort_values(by='h1n1_vaccine',ascending=False)
```

```python
from sklearn.metrics import classification_report
print(classification_report(train['h1n1_vaccine'],train['Predicted']))
```

```python
# Prediction on test data
test=pd.concat([x_test,y_test],axis=1)
test['Probability']=lgf.predict(x_test) 
test.head()
```

```python
test['Predicted']=np.where(test['Probability'] >= 0.7,1,0)  #thershold  >0.7
test.sort_values(by='h1n1_vaccine',ascending=False)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test['h1n1_vaccine'],test['Predicted']))
```

```python
# converting coefficients to odds
odds=np.exp(lgf.params)

# finding the percentage change 
perc_change_odds =(np.exp(lgf.params)-1)*100


# adding the odds to a dataframe
pd.DataFrame({'odds':odds,'change odd%':perc_change_odds},index=x_train.columns)
```
